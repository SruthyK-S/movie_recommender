```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("tmdb_movies_data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
movies = df[['id','original_title','overview','genres']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies['tags'] = movies['genres']+ ' ' + movies['overview'] 
```

```python
movies.head()
```

```python
movies = movies.drop(columns = ['overview','genres'])
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
movies.shape
```

```python
cv = CountVectorizer(max_features = 10866, stop_words = 'english')
```

```python
cv
```

```python
vector = cv.fit_transform(movies['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
movies.head(20)
```

```python
movies[movies['original_title']=="Furious 7"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[4])),reverse = True, key = lambda vector:vector[1])
for i in distance[0:5]:
    print(movies.iloc[i[0]].original_title)
```

```python
def recommend(movie):
    index = movies[movies['original_title']== movie].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse = True, key = lambda vector:vector[1])
    for i in distance[0:5]:
        print(movies.iloc[i[0]].original_title)
```

```python
recommend('Taken 3')
```
